```python
link=requests.get(link, headers=headers)
```

```python
soup = bs(link.content,'html')
```

```python
#print(soup.prettify())
```

```python
names = soup.find_all('span',class_='a-size-base-plus a-color-base a-text-normal')
names
```

```python
title = []
for i in range(0,len(names)):
    title.append(names[i].get_text())
title
```

```python
price_ = soup.find_all('span',class_='a-price-whole')
price_
```

```python
price = []
for i in range(0,len(price_)):
    price.append(price_[i].get_text())
price
```

```python
review = soup.find_all('span',class_='a-icon-alt')
review
```

```python
stars = []
for i in range(0,len(review)):
    stars.append(review[i].get_text())
stars
```

```python
img_ = soup.find_all('img',class_='s-image')
img_
```

```python
img = []
for i in range(0,len(img_)):
    img.append(img_[i].get('src'))
img
```

```python
product={'men_glasses':[]}
```

```python
for i in range(len(stars)):
    product['men_glasses'].append({'title':title[i],'price':price[i],'reveiw':stars[i],'image_link':img[i]})
```

```python
product
```

```python
with open('Data.json','w') as fp:
    json.dump(product, fp)
```
